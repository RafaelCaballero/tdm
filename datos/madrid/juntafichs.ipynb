```python
import pandas as pd
url_cont = "https://raw.githubusercontent.com/albercol/TFG/main/Fase%203/DatosLimpiosYOutliers/Contaminacion.csv"
url_met = "https://raw.githubusercontent.com/albercol/TFG/main/Fase%203/DatosLimpiosYOutliers/Meteorologia.csv"
url_traf = "https://raw.githubusercontent.com/albercol/TFG/main/Fase%203/DatosLimpiosYOutliers/Trafico.csv"
df_cont = pd.read_csv(url_cont)
df_met = pd.read_csv(url_met)
df_traf = pd.read_csv(url_traf)
df_contmet = pd.merge(df_cont, df_met, on="FECHAH",how="outer")
df_final = pd.merge(df_contmet, df_traf, on="FECHAH",how="outer")
df_final
```

```python
import urllib.request
festivos = pd.read_csv("https://raw.githubusercontent.com/RafaelCaballero/tdm/master/datos/calendario.csv",sep=",")

festivos.columns = ["FECHA","semana","festivo","TIPO","FESTIVIDAD"]
# los huecos en festivo parecen ser todos días laborables
festivos["festivo"].fillna("laborable", inplace=True)
festivos
```

```python
festivos.dropna(subset=["FECHA","semana"],inplace=True)
festivos["FECHA"] = pd.to_datetime(festivos['FECHA'])
festivos["festivo"] = festivos["festivo"].str.lower()
festivoscols = festivos[["FECHA","festivo"]]
# añadir la información de un dataframe al otro
festivoscols
```

```python
# left join
df_final["FECHA"] = pd.to_datetime(df_final['FECHA'])
df_final = df_final.merge(festivoscols, on='FECHA', how='left')
```

```python
df_final
```

```python
df_final["TIPO"] = df_final["festivo"]
df_final = df_final[['FECHAH', 'TIPO', 'CO', 'NO', 'NO2', 'PM2.5', 'PM10','NOx', 'O3',  
 'VELOCIDAD VIENTO', 'DIR. VIENTO', 'TEMPERATURA',
       'HUMEDAD RELATIVA', 'PRESION BAROMETRICA', 'PRECIPITACIONES', 
        'ES10', 'ES53', 'ESSUMA'
       ]]
df_final
```

```python
df_final.to_csv("../data/contmettraf.csv",index=False)
```
